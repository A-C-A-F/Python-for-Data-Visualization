# TASK #4. PLOT PIE CHART

```python
# how can I plot this pie chart?
```

```python
# specify first is I need to specify the values.
values = [20, 55, 5, 17, 3] # the values here includes the percentage.
# The second one is going to be my colors.
colors = ['g', 'r', 'y', 'b', 'm']
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']

# Use matplotlib to plot a pie chart 
plt.figure(figsize=(10,10))
plt.pie(values, colors=colors, labels = labels)
plt.title('STOCK PORTFOLIO')
```

what you guys
notice here is I have been able to plot a pie chart using map,
plot lib Here, you notice that Google has 55% is actually
more than half of my overall pie.
And then followed by here I have apple and then Tesla as well.
And then the smallest allocation will be for two amazon
and a TNT.


```python
# The explode. Use this if you are giving a presentation to leadership. 
#And you wanted to put some sort of emphasis on let's say some of these slices within my pie chart.

values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm']
explode = [0,0,0,0,0.2]
labels = ['AAPL', 'GOOG', 'T', 'TSLA', 'AMZN']

plt.figure(figsize=(10,10))
plt.pie(values, colors=colors, labels = labels, explode=explode)
plt.title('STOCK PORTFOLIO')
```

What we notice here is the amazon pie.
The actual slice has been exploded out of my full pie.
That's what explode attribute does. It just separates one of the slices here, out of my pie
if you wanted to for example show emphasis on a specific piece of information within my
pie chart.

# TASK #5. PLOT HISTOGRAMS

```python
daily_return_df
```

```python
# A histogram represents data using bars of various heights. 
# Each bar groups numbers into specific ranges. 
# Taller bars show that more data falls within that specific range.

mu = daily_return_df['FB'].mean() # mean of distribution
sigma = daily_return_df['FB'].std() # standard deviation of distribution

# standard deviation is the dispersion away from the mean
```

```python
# Plot Histograms
num_bins = 40
plt.figure(figsize=(15,9))
plt.hist(daily_return_df['FB'], num_bins, facecolor='blue')
plt.grid()
plt.title('Histogram: mu = ' + str(mu) + ', sigma= ' + str(sigma))
```

# TASK #9. SEABRON SCATTERPLOT & COUNTPLOT

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
# Import a more comprehensive dataset that can demonstrate the capacity of seaborn. 
# Import Cancer data drom the Sklearn library

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer

```

So this is the entire data.
The target is my output.
This dataset essentially has different
features from the breast cells and based on these features
like the cell sickness like the cells smoothness basically
doctors can predict whether these cells are cancerous or not,
whether they are malignant or benign basically.
And here these are the two different classes.
So either zero or one indicating if the if the actual data
point is um uh of that patient has cancer or not.


```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
plt.figure(figsize=(10,10))
sns.scatterplot(x='mean area',y='mean smoothness',data=df_cancer)
```

What we notice is all these data points
have been colored in the same color.
So I cannot tell which data points belong
to the malignant class and which data points belong
to the benign class to do that.
I can leverage my target column.


```python
# Plot scatter plot between mean area and mean smoothness with hue
plt.figure(figsize=(10,10))
sns.scatterplot(x='mean area',y='mean smoothness', hue = 'target', data=df_cancer)
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
plt.figure(figsize=(10,10))
sns.countplot(df_cancer['target'], label = 'Count')
```

# TASK #10. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS


```python
# Plot the pairplot
sns.pairplot(df_cancer, hue='target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'])
```

A pairplot plot a combination of every single variable that we have in our data. What pair plot also show us. It shows here on the diagonal elements. It can show us the actual distribution of my data.
So you can show me the distribution of both classes of the green class here and the blue class as well.
And these are again the malignant and benign cases in my cancer dataset.



The next plot that I wanted to do is correlation
plot. So simply I wanted to see how these variables
that we have, like the mean radius, mean perimeter, mean area
and so on.
How all these parameters or features are correlated
with each other.


```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize=(30,30))
sns.heatmap(df_cancer.corr(), annot = True)
```

Now we have been able to see all the different
correlations between all my features that we have in my data.
It's a little bit hard to see but essentially just
based based on this color code here, I can tell how
correlated these features are.
So for example, the the light color here indicates that this
is one correlation of one means we are perfectly correlated
with each other and you can see that of course here
in diagonal elements, all of the diagonal elements are ones
which makes sense because mean radius is perfectly correlated
with the main radius with itself.
However, here if you go to the negative values that the dark
cells here indicates that these cells are simply have inverse
correlation.
So it has correlation coefficient of -1.
So you can infer a lot of information about this graph.


```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. Think of it as distribution of my data.
sns.distplot(df_cancer['mean radius'], bins = 25, color = 'blue')

```

Now we have been able to plot our disk plot or distribution
